```python
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
tips=sb.load_dataset('tips')
tips.head()
```

```python
sb.regplot(x=list(range(100)),#regplot - регулярное построение РЕГРЕСИИ
           y=0.01*np.arange(100)+np.random.rand(100)) # случайные
           #значения по Y 
```

```python
# lmplot -- построение, принимающие dataframe
sb.lmplot(x="total_bill", y="tip", data=tips)
```

```python
sb.regplot(x='size', y='total_bill', data=tips)
```

```python
sb.regplot(x='size', y='tip', data=tips)
```

```python
sb.regplot(x='size',
           y='total_bill',
           x_jitter=0.2, # не меняет истинного значения(рассеяние)
           data=tips)
```

```python
sb.set_style('whitegrid')
sb.regplot(x='size',
           y='total_bill',
           x_estimator=np.mean,
           data=tips)
```
